```python
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import numpy as np

def si_f(t, y, beta, mu):
    s, i = y
    return np.array([-beta * i * s,
                     -mu * i + beta * i * s])

# parameters 
beta = 0.0075
mu = 5

sol= solve_ivp(si_f, [0, 5], [2000, 1], 
                rtol=1e-6, args=(beta, mu), method = 'RK45')


fig = plt.figure(); ax = fig.gca()
curves = ax.plot(sol.t, sol.y.T)
ax.legend(curves, ['S', 'I']);                                     
```
